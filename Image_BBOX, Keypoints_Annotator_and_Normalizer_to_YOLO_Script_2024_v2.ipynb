```python
import math
import os.path
import cv2

#  Customizable parameters
label_path = R'labels'            # The folder containing the images to be annotated
image_path = R'images'            # The folder where the annotation is saved to
key_point_num = 3                 # Number of key points
box_thickness = 1                 # frame thickness
small_box_thickness = 1           # The 8-point thickness of the box
label_thickness = 1               # The weight of the category font above the box
label_fontScale = 0.4             # Multiple of the category font above the box
key_thick = -1                    # Thickness of key points
key_text_thick = 1                # Text thickness on key points
key_text_scale = 0.4              # Magnification of text on key points
key_radius = 2                    # Key point drawing radius
dot = 6                           # Choose how many decimal places to keep
key_point_is = True               # Whether to mark key points. Set to None to mark ordinary yolo labels.
current_class = 0 

###---------------------- For ALL CLASSES -----------------------###

# # key colors
key_color = {
    0: (0, 0, 200),
    1: (255, 0, 0),
    2: (0, 222, 0),
    3: (255, 255, 0)} 

# key points text color
key_text_color = {
    0: (0, 100, 200),
    1: (255, 0, 0),
    2: (0, 255, 125),
    3: (0, 150, 128)} 

# bounding box color
box_color = {
    0: (0, 255, 0),
    1: (255, 0, 255),
    2: (128, 128, 0),
    3: (0, 128, 128)}  

# class labels
my_cls = {
    0: 'Player',
    1: 'Referee',
    2: 'Goalkeeper',
    3: 'Football'}  


# # Define dictionaries for each class
# my_cls = {
#     0: {
#         'key_color': (0, 0, 200),
#         'key_text_color': (0, 100, 200),
#         'box_color': (0, 255, 0),
#         'label': 'Player'
#     },
#     1: {
#         'key_color': (200, 0, 200),
#         'key_text_color': (0, 255, 0),
#         'box_color': (255, 255, 0),
#         'label': 'Referee'
#     },
#     2: {
#         'key_color': (125, 0, 200),
#         'key_text_color': (200, 100, 0),
#         'box_color': (0, 125, 225),
#         'label': 'Goalkeeper'
#     },
#     3: {
#         'key_color': (125, 120, 225),
#         'key_text_color': (200, 125, 0),
#         'box_color': (255, 150, 125),
#         'label': 'Football'
#     }
# }

# # Extract box_color dictionary for convenience
# key_color = {class_id: class_info['key_color'] for class_id, class_info in my_cls.items()}
# box_color = {class_id: class_info['box_color'] for class_id, class_info in my_cls.items()}
# key_text_color = {class_id: class_info['key_text_color'] for class_id, class_info in my_cls.items()}

# The default name of the box
final_class = {
    i: my_cls[i] if i in my_cls else str(i) for i in range(5)
}  

# Do not modify parameters
position = None  
label = None  
img = None  
Mouse_move = None  
label_index = None  
label_index_pos = None  
Mouse_insert = None  
draw_rectangle = None  
end_draw_rectangle = None  
append_str_temp = None  
empty_label = None 
# Parameters related to key points
key_points = None
key_points_move = None
key_x = None
key_y = None
key_v = None
key_box = None
box_move = None
key_insert = None
la_path = None

def flag_init():
    #  Initialize parameters
    global position, label, img, Mouse_insert, Mouse_move, label_index, draw_rectangle, end_draw_rectangle, append_str_temp, empty_label, \
        label_index_pos, key_v, key_x, key_y, key_x, key_box, key_points_move, key_points, box_move
    position = None
    Mouse_move = None
    label_index = None
    label_index_pos = None
    Mouse_insert = None
    draw_rectangle = None
    end_draw_rectangle = None
    append_str_temp = None
    empty_label = None
    key_points = None
    key_points_move = None
    key_x = None
    key_y = None
    key_v = None
    key_box = None
    box_move = None

# Used to draw small filled rectangles
def draw_rect_box(img, center, length_1, color=(0, 0, 255)):
    x1, y1 = center[0] - length_1, center[1] - length_1
    x2, y2 = center[0] + length_1, center[1] + length_1
    cv2.rectangle(img, (x1, y1), (x2, y2), color, thickness=-1)

# Used to read local images
def img_read(img_path, scale_):
    # scale_Fill in the minimum size of the screen
    image = cv2.imread(img_path)
    scale_x, scale_y, _ = image.shape
    if max(scale_x, scale_y) > scale_:
        scale = max(scale_x, scale_y) / scale_
        image = cv2.resize(image, (int(image.shape[1] / scale), int(image.shape[0] / scale)))
    return image

# Determine the distance between two points to determine whether the mouse position has entered the area of 8 points.
def distance(p1, p2):
    if math.sqrt((p2[0] - p1[0]) ** 2 + (p2[1] - p1[1]) ** 2) < 20:
        return True
    else:
        return False

# Draw a dotted rectangular frame. When switching to delete, the solid frame will be converted into a dotted frame.
def draw_dotted_rectangle(img, pt1, pt2, length_1=5, gap=6, thick=1, color=(100, 254, 100)):
    (x1, y1), (x2, y2) = pt1, pt2
    temp1, temp2 = x1, y1
    while x1 + length_1 < x2:
        cv2.line(img, (x1, y1), (x1 + length_1, y1), color, thickness=thick)
        cv2.line(img, (x1, y2), (x1 + length_1, y2), color, thickness=thick)
        x1 += length_1 + gap
    while y1 + length_1 < y2:
        cv2.line(img, (temp1, y1), (temp1, y1 + length_1), color, thickness=thick)
        cv2.line(img, (x1, y1), (x1, y1 + length_1), color, thickness=thick)
        y1 += length_1 + gap

# Display local tags into images
def label_show(img1, label_path, index):
    global small_box_thickness, box_thickness, label_fontScale, label_thickness, key_point_is, key_points, \
        key_radius, key_color, key_thick, key_text_scale, key_text_thick, key_text_color, label
    with open(la_path) as f:
        label = f.readlines()
    if len(label) == 0:
        return
    
    for i, points in enumerate(label):
        if key_point_is:
            # Get key point parameters
            key_points = points.split(' ')[5:]
        points = points.split(' ')[0:5]
        classify = int(float(points[0]))
        points.pop(0)
        point = [float(s.strip('\n')) for s in points]
        # point = list(map(float, points))
        scale_y, scale_x, _ = img1.shape
        x, y, w, h = int((point[0] - point[2] / 2) * scale_x), int(
            (point[1] - point[3] / 2) * scale_y), int(
            point[2] * scale_x), int(point[3] * scale_y)
        if i == index:
            cv2.rectangle(img1, (x, y), (x + w, y + h), (0, 255, 0), thickness=box_thickness) 
            #draw_dotted_rectangle(img1, (x, y), (x + w, y + h), length_1=5, gap=6, thick=1, color=(0, 255, 0))
            #draw_dotted_rectangle(img1, (x, y), (x + w, y + h), box_thickness)
        else:
            cv2.rectangle(img1, (x, y), (x + w, y + h), box_color[classify], thickness=box_thickness)
        # Draw the center point on the edge and four vertices
        draw_rect_box(img1, (x, int(0.5 * (y + y + h))), length_1=small_box_thickness)
        draw_rect_box(img1, (x + w - 1, int(0.5 * (y + y + h))), length_1=small_box_thickness)
        draw_rect_box(img1, (int(0.5 * (x + x + w)), y), length_1=3)
        draw_rect_box(img1, (int(0.5 * (x + x + w)), y + h), length_1=small_box_thickness)
        draw_rect_box(img1, (x, y), length_1=small_box_thickness)
        draw_rect_box(img1, (x + w, y), length_1=small_box_thickness)
        draw_rect_box(img1, (x + w, y + h), length_1=small_box_thickness)
        draw_rect_box(img1, (x, y + h), length_1=small_box_thickness)
        cv2.putText(img1, str(final_class[classify]), (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, label_fontScale,
                    (255, 0, 255), label_thickness)
        if key_point_is:
            # Get each key point in turn
            key_x = [float(i) for i in key_points[::3]]
            key_y = [float(i) for i in key_points[1::3]]
            key_v = [int(float(i)) for i in key_points[2::3]]
            index = 0
            key_point = zip(key_x, key_y)
            for p in key_point:
                cv2.circle(img, (int(p[0] * scale_x), int(p[1] * scale_y)), key_radius, key_color[key_v[index]],
                           thickness=key_thick,
                           lineType=cv2.LINE_AA)
                cv2.putText(img, str(index), (int(p[0] * scale_x - 5), int(p[1] * scale_y - 10)),
                            cv2.FONT_HERSHEY_SIMPLEX,
                            key_text_scale, key_text_color[key_v[index]], key_text_thick)
                index += 1
            key_points = None


# Callback function, used to record mouse operations
def mouse_event(event, x, y, flag, param):
    global label, img, position, Mouse_move, label_index, label_index_pos, dot, Mouse_insert, draw_rectangle, \
        end_draw_rectangle, key_points, key_v, key_x, key_y, key_x, key_box, key_points_move, box_move, key_insert, label_path
    scale_y, scale_x, _ = img.shape
    
    with open(la_path) as f:
        label = f.readlines()
    if key_insert is None and Mouse_insert is None and empty_label is None and event == cv2.EVENT_MOUSEMOVE and img is not None and label is not None and \
            Mouse_move is None:
        for i, la in enumerate(label):
            la = la.strip('\n').split(' ')
            if key_point_is:
                key_points = list(map(float, la))[5:]
            la = list(map(float, la))[0:5]
            x1, y1 = int((la[1] - la[3] / 2) * scale_x), int((la[2] - la[4] / 2) * scale_y)
            x2, y2 = x1 + int(la[3] * scale_x), y1 + int(la[4] * scale_y)
            # Here we determine which point the mouse has been placed on, so that we can make calculations when moving later.
            if distance((x, y), (x1, y1)):
                label_index_pos = 0
                position = (x, y)
                label_index = i
                box_move = True
                key_points_move = None
                break
            elif distance((x, y), (x2, y2)):
                label_index_pos = 1
                position = (x, y)
                label_index = i
                box_move = True
                key_points_move = None
                break
            elif distance((x, y), (x1, int(0.5 * y1 + 0.5 * y2))):
                label_index_pos = 2
                position = (x, y)
                label_index = i
                box_move = True
                key_points_move = None
                break
            elif distance((x, y), (int((x1 + x2) / 2), y2)):
                label_index_pos = 3
                position = (x, y)
                label_index = i
                box_move = True
                key_points_move = None
                break
            elif distance((x, y), (int((x1 + x2) / 2), y1)):
                label_index_pos = 4
                position = (x, y)
                label_index = i
                box_move = True
                key_points_move = None
                break
            elif distance((x, y), (x2, int(0.5 * y1 + 0.5 * y2))):
                label_index_pos = 5
                position = (x, y)
                label_index = i
                box_move = True
                key_points_move = None
                break
            elif distance((x, y), (x1, y2)):
                label_index_pos = 6
                position = (x, y)
                label_index = i
                box_move = True
                key_points_move = None
                break
            elif distance((x, y), (x2, y1)):
                label_index_pos = 7
                position = (x, y)
                label_index = i
                box_move = True
                key_points_move = None
                break
            else:
                label_index_pos = None
                position = None
                label_index = None
            if key_point_is:
                # Determine whether the mouse is placed on a key point
                key_x = [float(i) for i in key_points[::3]]
                key_y = [float(i) for i in key_points[1::3]]
                key_v = [float(i) for i in key_points[2::3]]  # visibility
                if len(key_x) == len(key_v) and len(key_x) == len(key_y):
                    for index, key_ in enumerate(key_x):
                        if distance((x, y), (int(key_ * scale_x), int(key_y[index] * scale_y))):
                            position = (x, y)
                            label_index, label_index_pos = i, index
                            key_box = la
                            key_points_move = True
                            box_move = None
                            break

    #  The comments from here to the next are all preparations for moving existing boxes.
    if position is not None and event == cv2.EVENT_LBUTTONDOWN:
        Mouse_move = True
        position = None
        
    if Mouse_move and box_move:
   
        temp_label = label[label_index]
        label.pop(label_index)
        temp_label = temp_label.strip('\n').split(' ')
        temp_label = [float(i) for i in temp_label]
        x_1, y_1 = (temp_label[1] - 0.5 * temp_label[3]), (temp_label[2] - 0.5 * temp_label[4])
        x_2, y_2 = x_1 + temp_label[3], y_1 + temp_label[4]
        # Determine which of the 8 points is moved
        if label_index_pos == 0:
            x_1, y_1 = x / scale_x, y / scale_y
        elif label_index_pos == 1:
            x_2, y_2 = x / scale_x, y / scale_y
        elif label_index_pos == 2:
            x_1 = x / scale_x
        elif label_index_pos == 3:
            y_2 = y / scale_y
        elif label_index_pos == 4:
            y_1 = y / scale_y
        elif label_index_pos == 5:
            x_2 = x / scale_x
        elif label_index_pos == 6:
            x_1, y_2 = x / scale_x, y / scale_y
        elif label_index_pos == 7:
            y_1, x_2 = y / scale_y, x / scale_x
        # Save the moved point information and add it to the label to create the effect of dynamically drawing a box.
        temp_label[0], temp_label[1], temp_label[2], temp_label[3], temp_label[4] = str(
            round((int(temp_label[0])), dot)), \
            str(round(((x_1 + x_2) * 0.5), dot)), str(round(((y_1 + y_2) * 0.5), dot)), str(
            round((abs(x_1 - x_2)), dot)), str(round((abs(y_1 - y_2)), dot))
        temp_label = [str(i) for i in temp_label]
        str_temp = ' '.join(temp_label) + '\n'
        label.append(str_temp)
        label_index = len(label) - 1
    elif Mouse_move and key_points_move:
        label.pop(label_index)
        key_x[label_index_pos] = round(x / scale_x, dot)
        key_y[label_index_pos] = round(y / scale_y, dot)
        key_box[0] = int(key_box[0])
        str_temp = ' '.join([str(j) for j in key_box])
        for index, kx in enumerate(key_x):
            str_temp += ' ' + str(kx) + ' ' + str(key_y[index]) + ' ' + str(int(key_v[index]))
        label.append(str_temp)
        label_index = len(label) - 1

    if Mouse_move and event == cv2.EVENT_LBUTTONUP:
        flag_init()

    # Here is the box for deletion
    if key_point_is is None and Mouse_insert is None and position is not None and event == cv2.EVENT_LBUTTONDBLCLK and Mouse_move is None:
        Mouse_insert = label_index

    if key_point_is and event == cv2.EVENT_LBUTTONDBLCLK and Mouse_move is None and key_points_move and box_move is None:
        key_insert = label_index_pos

    # Here is to add a new box
    if key_point_is is None and Mouse_insert is None and position is None and Mouse_move is None and event == cv2.EVENT_LBUTTONDOWN and end_draw_rectangle is None:
        draw_rectangle = [(x, y), (x, y)]

    # If the left mouse button is not released, the position of the second point will be continuously updated.
    elif Mouse_insert is None and draw_rectangle is not None and event == cv2.EVENT_MOUSEMOVE and end_draw_rectangle is None:
        draw_rectangle[1] = (x, y)

    elif Mouse_insert is None and draw_rectangle is not None and event == cv2.EVENT_LBUTTONUP:
        if end_draw_rectangle is None:
            draw_rectangle[1] = (x, y)
        if not distance(draw_rectangle[0], draw_rectangle[1]):
            end_draw_rectangle = True
        else:
            draw_rectangle = None

def create_file_key(img_path, label_path):
    empty_la = None
    if not os.path.exists(label_path):
        with open(label_path, 'w') as f:
            pass
        empty_la = True
    with open(label_path) as f:
        label_ = f.readlines()
    if len(label_) == 0 or label_[0] == '\n':
        empty_la = True
    # The 2 here is to reduce the original image to one-half
    img_s = img_read(img_path, 950)
    if key_point_is and empty_la:
        box_create = '0 0.5 0.5 0.3 0.3 '
        len_t = img_s.shape[1] // key_point_num
        key_num_x = [str(round((i * len_t + 20) / img_s.shape[1], dot)) + ' ' + str(0.5) + ' ' + '2' for i in
                     range(key_point_num)]
        with open(label_path, 'w') as f:
            f.write(box_create + ' '.join(key_num_x))

# def main(img_path, label_path):
#     global img, position, label, Mouse_insert, draw_rectangle, end_draw_rectangle, append_str_temp, empty_label, \
#         Mouse_move, dot, box_move, key_insert

#     if not os.path.exists(label_path):
#         empty_label = True
#         with open(label_path, 'w') as f:
#             pass
#     with open(label_path) as f:
#         label = f.readlines()
#     if len(label) == 0 or label[0] == '\n':
#         empty_label = True
#     # The 2 here is to reduce the original image to one-half
#     img_s = img_read(img_path, 950)
#     if key_point_is and empty_label:
#         box_create = '0 0.5 0.5 0.3 0.3 '
#         len_t = img_s.shape[1] // key_point_num
#         key_num_x = [str(round((i * len_t + 20) / img_s.shape[1], dot)) + ' ' + str(0.5) + ' ' + '2' for i in
#                      range(key_point_num)]
#         with open(label_path, 'w') as f:
#             f.write(box_create + ' '.join(key_num_x))
#             label = box_create + ' '.join(key_num_x)
#     # Create a callback function and bind the window
#     cv2.namedWindow('image', cv2.WINDOW_NORMAL)
#     cv2.resizeWindow('image', img_s.shape[1], img_s.shape[0])
#     cv2.setMouseCallback('image', mouse_event)
#     # Where to refresh the image
#     while True:
#         # First read the tag to initialize the display
#         with open(label_path, 'w') as f:
#             for i in label:
#                 f.write(i)
#         # If the mouse selects one of the 8 points of the box, draw an empty circle around the mouse
#         if Mouse_insert is None and draw_rectangle is None and position is not None and key_insert is None:
#             img = img_s.copy()
#             label_show(img, label_path, Mouse_insert)
#             cv2.circle(img, position, 10, (0, 255, 100), 2)
#         # If you choose to start adding a new box, the box formed between the mouse starting 
#         # point and the movement process will be continuously drawn.
#         elif draw_rectangle is not None and end_draw_rectangle is None:
#             img = img_s.copy()
#             label_show(img, label_path, Mouse_insert)
#             cv2.rectangle(img, draw_rectangle[0], draw_rectangle[1], color=box_color[1], thickness=2)
#         # When the mouse is released, the two point positions are recorded and prompted to select a category.
#         elif draw_rectangle is not None and end_draw_rectangle:
#             scale_y, scale_x, _ = img.shape
#             x1, y1 = draw_rectangle[0]
#             x2, y2 = draw_rectangle[1]
#             w1, h1 = abs(x2 - x1), abs(y2 - y1)
#             append_str_temp = str(round((x1 + x2) / 2 / scale_x, dot)) + ' ' + str(
#                 round((y1 + y2) / 2 / scale_y, dot)) + ' ' + \
#                               str(round((w1 / scale_x), dot)) + ' ' + str(round((h1 / scale_y), dot)) + '\n'
#             cv2.putText(img, 'choose your classify', (0, img.shape[0] // 2 - 30),
#                         cv2.FONT_HERSHEY_SIMPLEX,
#                         0.7, (255, 0, 255), 2)
#             cv2.putText(img, ' '.join([str(i) + ':' + my_cls[i] for i in my_cls]), (0, img.shape[0] // 2 + 30),
#                         cv2.FONT_HERSHEY_SIMPLEX,
#                         0.7, (100, 255, 255), 2)
#         elif key_insert is not None:
#             position, Mouse_move, box_move = None, None, None  # Disable other operations
#             cv2.putText(img, 'Switching visibility: 0     1    2', (0, img.shape[0] // 2 - 30),
#                         cv2.FONT_HERSHEY_SIMPLEX,
#                         1, (100, 255, 255), 2,
#                         lineType=cv2.LINE_AA)
#         # If there is no logo, a picture will be displayed normally.
#         else:
#             img = img_s.copy()
#             if Mouse_insert is not None:
#                 position, Mouse_move = None, None
#                 cv2.putText(img, 'delete: W, exit: E', (0, img.shape[0] // 2 - 30),
#                             cv2.FONT_HERSHEY_SIMPLEX,
#                             1, (100, 255, 255), 2,
#                             lineType=cv2.LINE_AA)
#             label_show(img, label_path, Mouse_insert)
#         cv2.imshow('image', img)

#         # key is used to obtain keyboard input
#         key = cv2.waitKey(1)
#         # Enter Q to exit
#         if key == ord('Q'):
#             # Exit button
#             break
#         # If the key input is W, delete the selected box
#         if Mouse_insert is not None and key == ord('W'):
#             label.pop(Mouse_insert)
#             Mouse_insert = None
#         elif key_insert is not None and key == ord('0'):
#             with open(label_path, 'r') as f:
#                 label_temp = f.read()
#                 str_temp = label_temp.split(' ')
#                 str_temp[3 * int(key_insert) + 7] = '0'
#                 str_temp[3 * int(key_insert) + 7 - 1] = '0'
#                 str_temp[3 * int(key_insert) + 7 - 2] = '0'
#             with open(label_path, 'w') as f:
#                 f.write(' '.join(str_temp))
#                 label = ' '.join(str_temp)
#                 key_insert = None
#         elif key_insert is not None and key == ord('1'):
#             with open(label_path, 'r') as f:
#                 label_temp = f.read()
#                 str_temp = label_temp.split(' ')
#                 str_temp[3 * int(key_insert) + 7] = '1'
#             with open(label_path, 'w') as f:
#                 f.write(' '.join(str_temp))
#                 label = ' '.join(str_temp)
#                 key_insert = None
#         elif key_insert is not None and key == ord('2'):
#             with open(label_path, 'r') as f:
#                 label_temp = f.read()
#                 str_temp = label_temp.split(' ')
#                 str_temp[3 * int(key_insert) + 7] = '2'
#             with open(label_path, 'w') as f:
#                 f.write(' '.join(str_temp))
#                 label = ' '.join(str_temp)
#                 key_insert = None
#         # If the input is E, exit from the checked box state.
#         elif key == ord('E'):
#             Mouse_insert = None
#         # Get the category of input via keyboard
#         elif Mouse_move is None and Mouse_insert is None and draw_rectangle is not None and end_draw_rectangle is not None \
#                 and 48 <= key <= 57:
#             str_temp = str(chr(key)) + ' ' + append_str_temp
#             label.append(str_temp)
#             append_str_temp, draw_rectangle, end_draw_rectangle, empty_label = None, None, None, None

def switch_class():
    global current_class
    current_class = (current_class + 1) % len(my_cls)  # Cycle through classes    

def main(img_path, label_path):
    global img, position, label, Mouse_insert, draw_rectangle, end_draw_rectangle, append_str_temp, empty_label, \
        Mouse_move, dot, box_move, key_insert, current_class

    if not os.path.exists(label_path):
        empty_label = True
        with open(label_path, 'w'):
            pass
    with open(label_path) as f:
        label = f.readlines()
    if len(label) == 0 or label[0] == '\n':
        empty_label = True

    img_s = img_read(img_path, 950)
    if key_point_is and empty_label:
        box_create = '0 0.5 0.5 0.3 0.3 '
        len_t = img_s.shape[1] // key_point_num
        key_num_x = [str(round((i * len_t + 20) / img_s.shape[1], dot)) + ' ' + str(0.5) + ' ' + '2' for i in
                     range(key_point_num)]
        with open(label_path, 'w') as f:
            f.write(box_create + ' '.join(key_num_x))
            label = box_create + ' '.join(key_num_x)

    cv2.namedWindow('image', cv2.WINDOW_NORMAL)
    cv2.resizeWindow('image', img_s.shape[1], img_s.shape[0])
    cv2.setMouseCallback('image', mouse_event)

    while True:
        with open(label_path, 'w') as f:
            for i in label:
                f.write(i)

        if Mouse_insert is None and draw_rectangle is None and position is not None and key_insert is None:
            img = img_s.copy()
            label_show(img, label_path, Mouse_insert)
            cv2.circle(img, position, 10, (0, 255, 100), 2)

        elif draw_rectangle is not None and end_draw_rectangle is None:
            img = img_s.copy()
            label_show(img, label_path, Mouse_insert)
            cv2.rectangle(img, draw_rectangle[0], draw_rectangle[1], color=box_color[current_class], thickness=2)

        elif draw_rectangle is not None and end_draw_rectangle:
            scale_y, scale_x, _ = img.shape
            x1, y1 = draw_rectangle[0]
            x2, y2 = draw_rectangle[1]
            w1, h1 = abs(x2 - x1), abs(y2 - y1)
            append_str_temp = (
                    str(round((x1 + x2) / 2 / scale_x, dot)) + ' ' +
                    str(round((y1 + y2) / 2 / scale_y, dot)) + ' ' +
                    str(round((w1 / scale_x), dot)) + ' ' +
                    str(round((h1 / scale_y), dot)) + ' ' +
                    str(current_class) + '\n'
            )
            cv2.putText(
                img,
                f'Selected Class: {current_class} (Press A/D to switch class)',
                (0, img.shape[0] // 2 + 30),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.7, (100, 255, 255), 2
            )

        elif key_insert is not None:
            position, Mouse_move, box_move = None, None, None
            cv2.putText(
                img,
                'Switching visibility: 0     1    2 (Press A/D to switch class)',
                (0, img.shape[0] // 2 - 30),
                cv2.FONT_HERSHEY_SIMPLEX,
                1, (100, 255, 255), 2,
                lineType=cv2.LINE_AA
            )
        else:
            img = img_s.copy()
            if Mouse_insert is not None:
                position, Mouse_move = None, None
                cv2.putText(
                    img,
                    'delete: W, exit: E, switch class: S (Press A/D to switch class)',
                    (0, img.shape[0] // 2 - 30),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    1, (100, 255, 255), 2,
                    lineType=cv2.LINE_AA
                )
            label_show(img, label_path, Mouse_insert)

        cv2.imshow('image', img)

        key = cv2.waitKey(1)

        if key == ord('Q'):
            break
        elif key == ord('E'):
            position, Mouse_move, box_move, end_draw_rectangle, append_str_temp = None, None, None, None, None
            if draw_rectangle is not None:
                draw_rectangle = None
            else:
                try:
                    label.pop()
                except IndexError:
                    pass
        elif key == ord('W'):
            position, Mouse_move, box_move, end_draw_rectangle, append_str_temp = None, None, None, None, None
            if draw_rectangle is not None:
                draw_rectangle = None
            else:
                try:
                    label.pop()
                except IndexError:
                    pass
        elif key == ord('S'):
            current_class = (current_class + 1) % len(my_cls)
        elif key == ord('A'):
            current_class = (current_class - 1) % len(my_cls)
        elif key == ord('D'):
            current_class = (current_class + 1) % len(my_cls)
        elif key == 27:
            break

    cv2.destroyAllWindows()

#         elif draw_rectangle is not None and end_draw_rectangle:
#             scale_y, scale_x, _ = img.shape
#             x1, y1 = draw_rectangle[0]
#             x2, y2 = draw_rectangle[1]
#             w1, h1 = abs(x2 - x1), abs(y2 - y1)
#             append_str_temp = str(round((x1 + x2) / 2 / scale_x, dot)) + ' ' + str(
#                 round((y1 + y2) / 2 / scale_y, dot)) + ' ' + \
#                               str(round((w1 / scale_x), dot)) + ' ' + str(round((h1 / scale_y), dot)) + '\n'
#             cv2.putText(img, 'choose your classify (Press S to switch class)', (0, img.shape[0] // 2 - 30),
#                         cv2.FONT_HERSHEY_SIMPLEX,
#                         0.7, (255, 0, 255), 2)
#             cv2.putText(img, ' '.join([str(i) + ':' + my_cls[i] for i in my_cls]), (0, img.shape[0] // 2 + 30),
#                         cv2.FONT_HERSHEY_SIMPLEX,
#                         0.7, (100, 255, 255), 2)
#         elif key_insert is not None:
#             position, Mouse_move, box_move = None, None, None
#             cv2.putText(img, 'Switching visibility: 0     1    2', (0, img.shape[0] // 2 - 30),
#                         cv2.FONT_HERSHEY_SIMPLEX,
#                         1, (100, 255, 255), 2,
#                         lineType=cv2.LINE_AA)
#         else:
#             img = img_s.copy()
#             if Mouse_insert is not None:
#                 position, Mouse_move = None, None
#                 cv2.putText(img, 'delete: W, exit: E, switch class: S', (0, img.shape[0] // 2 - 30),
#                             cv2.FONT_HERSHEY_SIMPLEX,
#                             1, (100, 255, 255), 2,
#                             lineType=cv2.LINE_AA)
#             label_show(img, label_path, Mouse_insert)

#         cv2.imshow('image', img)

#         key = cv2.waitKey(1)

#         if key == ord('Q'):
#             break
#         elif key == ord('E'):
#             position, Mouse_move, box_move, end_draw_rectangle, append_str_temp = None, None, None, None, None
#             if draw_rectangle is not None:
#                 draw_rectangle = None
#             else:
#                 try:
#                     label.pop()
#                 except IndexError:
#                     pass
#         elif key == ord('W'):
#             position, Mouse_move, box_move, end_draw_rectangle, append_str_temp = None, None, None, None, None
#             if draw_rectangle is not None:
#                 draw_rectangle = None
#             else:
#                 try:
#                     label.pop()
#                 except IndexError:
#                     pass
#         elif key == ord('S'):
#             current_class = (current_class + 1) % len(my_cls)
#         elif key == ord('A'):
#             position, Mouse_move, box_move, end_draw_rectangle, append_str_temp = None, None, None, None, None
#             if draw_rectangle is not None:
#                 draw_rectangle = None
#             else:
#                 try:
#                     label.pop()
#                 except IndexError:
#                     pass
#         elif key == ord('D'):
#             position, Mouse_move, box_move, end_draw_rectangle, append_str_temp = None, None, None, None, None
#             if draw_rectangle is not None:
#                 draw_rectangle = None
#             else:
#                 try:
#                     label.pop()
#                 except IndexError:
#                     pass
#         elif key == 27:
#             break

#     cv2.destroyAllWindows()         
            
if __name__ == '__main__':
    global label_path_main, image_path_main
    image_ = os.listdir(image_path)
    for im in image_:
        flag_init()
        im_path = os.path.join(image_path, im)
        la_path = os.path.join(label_path, im.split('.')[0] + '.txt')
        if key_point_is:
            create_file_key(im_path, la_path)
        main(im_path, la_path)
```

```python
###-----------------------------------JAN 02, 2024----------------------------------------

import math
import os.path
import cv2

#  Customizable parameters
label_path = R'labels'            # The folder containing the images to be annotated
image_path = R'images'            # The folder where the annotation is saved to
key_point_num = 3                 # Number of key points
box_thickness = 1                 # frame thickness
small_box_thickness = 1           # The 8-point thickness of the box
label_thickness = 1               # The weight of the category font above the box
label_fontScale = 0.4             # Multiple of the category font above the box
key_thick = -1                    # Thickness of key points
key_text_thick = 1                # Text thickness on key points
key_text_scale = 0.4              # Magnification of text on key points
key_radius = 2                    # Key point drawing radius
dot = 6                           # Choose how many decimal places to keep
key_point_is = True               # Whether to mark key points. Set to None to mark ordinary yolo labels.
current_class = 0 

###---------------------- For ALL CLASSES -----------------------###

# # key colors
key_color = {
    0: (0, 0, 200),
    1: (255, 0, 0),
    2: (0, 222, 0),
    3: (255, 255, 0)} 

# key points text color
key_text_color = {
    0: (0, 100, 200),
    1: (255, 0, 0),
    2: (0, 255, 125),
    3: (0, 150, 128)} 

# bounding box color
box_color = {
    0: (0, 255, 0),
    1: (255, 0, 255),
    2: (128, 128, 0),
    3: (0, 128, 128)}  

# class labels
my_cls = {
    0: 'Player',
    1: 'Referee',
    2: 'Goalkeeper',
    3: 'Football'}  

# The default name of the box
final_class = {
    i: my_cls[i] if i in my_cls else str(i) for i in range(5)
}  

# Do not modify parameters
position = None  
label = None  
img = None  
Mouse_move = None  
label_index = None  
label_index_pos = None  
Mouse_insert = None  
draw_rectangle = None  
end_draw_rectangle = None  
append_str_temp = None  
empty_label = None 
# Parameters related to key points
key_points = None
key_points_move = None
key_x = None
key_y = None
key_v = None
key_box = None
box_move = None
key_insert = None
la_path = None

def flag_init():
    #  Initialize parameters
    global position, label, img, Mouse_insert, Mouse_move, label_index, draw_rectangle, end_draw_rectangle, append_str_temp, empty_label, \
        label_index_pos, key_v, key_x, key_y, key_x, key_box, key_points_move, key_points, box_move
    position = None
    Mouse_move = None
    label_index = None
    label_index_pos = None
    Mouse_insert = None
    draw_rectangle = None
    end_draw_rectangle = None
    append_str_temp = None
    empty_label = None
    key_points = None
    key_points_move = None
    key_x = None
    key_y = None
    key_v = None
    key_box = None
    box_move = None

# Used to draw small filled rectangles
def draw_rect_box(img, center, length_1, color=(0, 0, 255)):
    x1, y1 = center[0] - length_1, center[1] - length_1
    x2, y2 = center[0] + length_1, center[1] + length_1
    cv2.rectangle(img, (x1, y1), (x2, y2), color, thickness=-1)

# Used to read local images
def img_read(img_path, scale_):
    # scale_Fill in the minimum size of the screen
    image = cv2.imread(img_path)
    scale_x, scale_y, _ = image.shape
    if max(scale_x, scale_y) > scale_:
        scale = max(scale_x, scale_y) / scale_
        image = cv2.resize(image, (int(image.shape[1] / scale), int(image.shape[0] / scale)))
    return image

# Determine the distance between two points to determine whether the mouse position has entered the area of 8 points.
def distance(p1, p2):
    if math.sqrt((p2[0] - p1[0]) ** 2 + (p2[1] - p1[1]) ** 2) < 20:
        return True
    else:
        return False

# Draw a dotted rectangular frame. When switching to delete, the solid frame will be converted into a dotted frame.
def draw_dotted_rectangle(img, pt1, pt2, length_1=5, gap=6, thick=1, color=(100, 254, 100)):
    (x1, y1), (x2, y2) = pt1, pt2
    temp1, temp2 = x1, y1
    while x1 + length_1 < x2:
        cv2.line(img, (x1, y1), (x1 + length_1, y1), color, thickness=thick)
        cv2.line(img, (x1, y2), (x1 + length_1, y2), color, thickness=thick)
        x1 += length_1 + gap
    while y1 + length_1 < y2:
        cv2.line(img, (temp1, y1), (temp1, y1 + length_1), color, thickness=thick)
        cv2.line(img, (x1, y1), (x1, y1 + length_1), color, thickness=thick)
        y1 += length_1 + gap

# Display local tags into images
def label_show(img1, label_path, index):
    global small_box_thickness, box_thickness, label_fontScale, label_thickness, key_point_is, key_points, \
        key_radius, key_color, key_thick, key_text_scale, key_text_thick, key_text_color, label
    with open(la_path) as f:
        label = f.readlines()
    if len(label) == 0:
        return
    
    for i, points in enumerate(label):
        if key_point_is:
            # Get key point parameters
            key_points = points.split(' ')[5:]
        points = points.split(' ')[0:5]
        classify = int(float(points[0]))
        points.pop(0)
        point = [float(s.strip('\n')) for s in points]
        # point = list(map(float, points))
        scale_y, scale_x, _ = img1.shape
        x, y, w, h = int((point[0] - point[2] / 2) * scale_x), int(
            (point[1] - point[3] / 2) * scale_y), int(
            point[2] * scale_x), int(point[3] * scale_y)
        if i == index:
            cv2.rectangle(img1, (x, y), (x + w, y + h), (0, 255, 0), thickness=box_thickness) 
            #draw_dotted_rectangle(img1, (x, y), (x + w, y + h), length_1=5, gap=6, thick=1, color=(0, 255, 0))
            #draw_dotted_rectangle(img1, (x, y), (x + w, y + h), box_thickness)
        else:
            cv2.rectangle(img1, (x, y), (x + w, y + h), box_color[classify], thickness=box_thickness)
        # Draw the center point on the edge and four vertices
        draw_rect_box(img1, (x, int(0.5 * (y + y + h))), length_1=small_box_thickness)
        draw_rect_box(img1, (x + w - 1, int(0.5 * (y + y + h))), length_1=small_box_thickness)
        draw_rect_box(img1, (int(0.5 * (x + x + w)), y), length_1=3)
        draw_rect_box(img1, (int(0.5 * (x + x + w)), y + h), length_1=small_box_thickness)
        draw_rect_box(img1, (x, y), length_1=small_box_thickness)
        draw_rect_box(img1, (x + w, y), length_1=small_box_thickness)
        draw_rect_box(img1, (x + w, y + h), length_1=small_box_thickness)
        draw_rect_box(img1, (x, y + h), length_1=small_box_thickness)
        cv2.putText(img1, str(final_class[classify]), (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, label_fontScale,
                    (255, 0, 255), label_thickness)
        if key_point_is:
            # Get each key point in turn
            key_x = [float(i) for i in key_points[::3]]
            key_y = [float(i) for i in key_points[1::3]]
            key_v = [int(float(i)) for i in key_points[2::3]]
            index = 0
            key_point = zip(key_x, key_y)
            for p in key_point:
                cv2.circle(img, (int(p[0] * scale_x), int(p[1] * scale_y)), key_radius, key_color[key_v[index]],
                           thickness=key_thick,
                           lineType=cv2.LINE_AA)
                cv2.putText(img, str(index), (int(p[0] * scale_x - 5), int(p[1] * scale_y - 10)),
                            cv2.FONT_HERSHEY_SIMPLEX,
                            key_text_scale, key_text_color[key_v[index]], key_text_thick)
                index += 1
            key_points = None


# Callback function, used to record mouse operations
def mouse_event(event, x, y, flag, param):
    global label, img, position, Mouse_move, label_index, label_index_pos, dot, Mouse_insert, draw_rectangle, \
        end_draw_rectangle, key_points, key_v, key_x, key_y, key_x, key_box, key_points_move, box_move, key_insert, label_path
    scale_y, scale_x, _ = img.shape
    
    with open(la_path) as f:
        label = f.readlines()
    if key_insert is None and Mouse_insert is None and empty_label is None and event == cv2.EVENT_MOUSEMOVE and img is not None and label is not None and \
            Mouse_move is None:
        for i, la in enumerate(label):
            la = la.strip('\n').split(' ')
            if key_point_is:
                key_points = list(map(float, la))[5:]
            la = list(map(float, la))[0:5]
            x1, y1 = int((la[1] - la[3] / 2) * scale_x), int((la[2] - la[4] / 2) * scale_y)
            x2, y2 = x1 + int(la[3] * scale_x), y1 + int(la[4] * scale_y)
            # Here we determine which point the mouse has been placed on, so that we can make calculations when moving later.
            if distance((x, y), (x1, y1)):
                label_index_pos = 0
                position = (x, y)
                label_index = i
                box_move = True
                key_points_move = None
                break
            elif distance((x, y), (x2, y2)):
                label_index_pos = 1
                position = (x, y)
                label_index = i
                box_move = True
                key_points_move = None
                break
            elif distance((x, y), (x1, int(0.5 * y1 + 0.5 * y2))):
                label_index_pos = 2
                position = (x, y)
                label_index = i
                box_move = True
                key_points_move = None
                break
            elif distance((x, y), (int((x1 + x2) / 2), y2)):
                label_index_pos = 3
                position = (x, y)
                label_index = i
                box_move = True
                key_points_move = None
                break
            elif distance((x, y), (int((x1 + x2) / 2), y1)):
                label_index_pos = 4
                position = (x, y)
                label_index = i
                box_move = True
                key_points_move = None
                break
            elif distance((x, y), (x2, int(0.5 * y1 + 0.5 * y2))):
                label_index_pos = 5
                position = (x, y)
                label_index = i
                box_move = True
                key_points_move = None
                break
            elif distance((x, y), (x1, y2)):
                label_index_pos = 6
                position = (x, y)
                label_index = i
                box_move = True
                key_points_move = None
                break
            elif distance((x, y), (x2, y1)):
                label_index_pos = 7
                position = (x, y)
                label_index = i
                box_move = True
                key_points_move = None
                break
            else:
                label_index_pos = None
                position = None
                label_index = None
            if key_point_is:
                # Determine whether the mouse is placed on a key point
                key_x = [float(i) for i in key_points[::3]]
                key_y = [float(i) for i in key_points[1::3]]
                key_v = [float(i) for i in key_points[2::3]]  # visibility
                if len(key_x) == len(key_v) and len(key_x) == len(key_y):
                    for index, key_ in enumerate(key_x):
                        if distance((x, y), (int(key_ * scale_x), int(key_y[index] * scale_y))):
                            position = (x, y)
                            label_index, label_index_pos = i, index
                            key_box = la
                            key_points_move = True
                            box_move = None
                            break

    #  The comments from here to the next are all preparations for moving existing boxes.
    if position is not None and event == cv2.EVENT_LBUTTONDOWN:
        Mouse_move = True
        position = None
        
    if Mouse_move and box_move:
   
        temp_label = label[label_index]
        label.pop(label_index)
        temp_label = temp_label.strip('\n').split(' ')
        temp_label = [float(i) for i in temp_label]
        x_1, y_1 = (temp_label[1] - 0.5 * temp_label[3]), (temp_label[2] - 0.5 * temp_label[4])
        x_2, y_2 = x_1 + temp_label[3], y_1 + temp_label[4]
        # Determine which of the 8 points is moved
        if label_index_pos == 0:
            x_1, y_1 = x / scale_x, y / scale_y
        elif label_index_pos == 1:
            x_2, y_2 = x / scale_x, y / scale_y
        elif label_index_pos == 2:
            x_1 = x / scale_x
        elif label_index_pos == 3:
            y_2 = y / scale_y
        elif label_index_pos == 4:
            y_1 = y / scale_y
        elif label_index_pos == 5:
            x_2 = x / scale_x
        elif label_index_pos == 6:
            x_1, y_2 = x / scale_x, y / scale_y
        elif label_index_pos == 7:
            y_1, x_2 = y / scale_y, x / scale_x
        # Save the moved point information and add it to the label to create the effect of dynamically drawing a box.
        temp_label[0], temp_label[1], temp_label[2], temp_label[3], temp_label[4] = str(
            round((int(temp_label[0])), dot)), \
            str(round(((x_1 + x_2) * 0.5), dot)), str(round(((y_1 + y_2) * 0.5), dot)), str(
            round((abs(x_1 - x_2)), dot)), str(round((abs(y_1 - y_2)), dot))
        temp_label = [str(i) for i in temp_label]
        str_temp = ' '.join(temp_label) + '\n'
        label.append(str_temp)
        label_index = len(label) - 1
    elif Mouse_move and key_points_move:
        label.pop(label_index)
        key_x[label_index_pos] = round(x / scale_x, dot)
        key_y[label_index_pos] = round(y / scale_y, dot)
        key_box[0] = int(key_box[0])
        str_temp = ' '.join([str(j) for j in key_box])
        for index, kx in enumerate(key_x):
            str_temp += ' ' + str(kx) + ' ' + str(key_y[index]) + ' ' + str(int(key_v[index]))
        label.append(str_temp)
        label_index = len(label) - 1

    if Mouse_move and event == cv2.EVENT_LBUTTONUP:
        flag_init()

    # Here is the box for deletion
    if key_point_is is None and Mouse_insert is None and position is not None and event == cv2.EVENT_LBUTTONDBLCLK and Mouse_move is None:
        Mouse_insert = label_index

    if key_point_is and event == cv2.EVENT_LBUTTONDBLCLK and Mouse_move is None and key_points_move and box_move is None:
        key_insert = label_index_pos

    # Here is to add a new box
    if key_point_is is None and Mouse_insert is None and position is None and Mouse_move is None and event == cv2.EVENT_LBUTTONDOWN and end_draw_rectangle is None:
        draw_rectangle = [(x, y), (x, y)]

    # If the left mouse button is not released, the position of the second point will be continuously updated.
    elif Mouse_insert is None and draw_rectangle is not None and event == cv2.EVENT_MOUSEMOVE and end_draw_rectangle is None:
        draw_rectangle[1] = (x, y)

    elif Mouse_insert is None and draw_rectangle is not None and event == cv2.EVENT_LBUTTONUP:
        if end_draw_rectangle is None:
            draw_rectangle[1] = (x, y)
        if not distance(draw_rectangle[0], draw_rectangle[1]):
            end_draw_rectangle = True
        else:
            draw_rectangle = None

def create_file_key(img_path, label_path):
    empty_la = None
    if not os.path.exists(label_path):
        with open(label_path, 'w') as f:
            pass
        empty_la = True
    with open(label_path) as f:
        label_ = f.readlines()
    if len(label_) == 0 or label_[0] == '\n':
        empty_la = True
    # The 2 here is to reduce the original image to one-half
    img_s = img_read(img_path, 950)
    if key_point_is and empty_la:
        box_create = '0 0.5 0.5 0.3 0.3 '
        len_t = img_s.shape[1] // key_point_num
        key_num_x = [str(round((i * len_t + 20) / img_s.shape[1], dot)) + ' ' + str(0.5) + ' ' + '2' for i in
                     range(key_point_num)]
        with open(label_path, 'w') as f:
            f.write(box_create + ' '.join(key_num_x))

# def switch_class():
#     global current_class
#     current_class = (current_class + 1) % len(my_cls)  # Cycle through classes    

# def on_class_change(value):
#     global current_class
#     current_class = value

def switch_class():
    global current_class, img, position, label, Mouse_insert, draw_rectangle, end_draw_rectangle, append_str_temp, \
        empty_label, Mouse_move, dot, box_move, key_insert

    # Handle class switching on key press
    current_class += 1
    if current_class >= len(my_cls):
        current_class = 0

    if key_insert is not None:
        if end_draw_rectangle is None:
            end_draw_rectangle = True
        else:
            end_draw_rectangle = None

    if Mouse_insert is not None:
        position, Mouse_move = None, None
        cv2.putText(
            img,
            'delete: W, exit: E, switch class: S (Press A/D to switch class)',
            (0, img.shape[0] // 2 - 30),
            cv2.FONT_HERSHEY_SIMPLEX,
            1, (100, 255, 255), 2,
            lineType=cv2.LINE_AA
        )
    label_show(img, label_path, Mouse_insert)


def on_class_change(value):
    global current_class, img, position, label, Mouse_insert, draw_rectangle, end_draw_rectangle, append_str_temp, \
        empty_label, Mouse_move, dot, box_move, key_insert

    # Handle class switching from trackbar
    current_class = value

    if key_insert is not None:
        if end_draw_rectangle is None:
            end_draw_rectangle = True
        else:
            end_draw_rectangle = None

    if Mouse_insert is not None:
        position, Mouse_move = None, None
        cv2.putText(
            img,
            'delete: W, exit: E, switch class: S (Press A/D to switch class)',
            (0, img.shape[0] // 2 - 30),
            cv2.FONT_HERSHEY_SIMPLEX,
            1, (100, 255, 255), 2,
            lineType=cv2.LINE_AA
        )
    label_show(img, label_path, Mouse_insert)
    
def main(img_path, label_path):
    global img, position, label, Mouse_insert, draw_rectangle, end_draw_rectangle, append_str_temp, empty_label, \
        Mouse_move, dot, box_move, key_insert, current_class

    img_s = cv2.imread(img_path)

    # Reset the current class to the default class (0)
    current_class = 0
    

    cv2.namedWindow('image', cv2.WINDOW_NORMAL)
    cv2.resizeWindow('image', img_s.shape[1], img_s.shape[0])
    cv2.setMouseCallback('image', mouse_event)

    # Create the trackbar
    #cv2.createTrackbar('Class', 'image', current_class, len(my_cls) - 1, lambda x: None)
    cv2.createTrackbar('Class', 'image', current_class, len(my_cls) - 1, on_class_change)
    
    while True:
        if label is not None:  # Add a check for None
            with open(label_path, 'w') as f:
                for i in label:
                    f.write(i)

        if Mouse_insert is None and draw_rectangle is None and position is not None and key_insert is None:
            img = img_s.copy()
            label_show(img, label_path, Mouse_insert)
            cv2.circle(img, position, 10, (0, 255, 100), 2)

        elif draw_rectangle is not None and end_draw_rectangle is None:
            img = img_s.copy()
            label_show(img, label_path, Mouse_insert)
            cv2.rectangle(img, draw_rectangle[0], draw_rectangle[1], color=box_color[current_class], thickness=2)

        elif draw_rectangle is not None and end_draw_rectangle:
            scale_y, scale_x, _ = img.shape
            x1, y1 = draw_rectangle[0]
            x2, y2 = draw_rectangle[1]
            w1, h1 = abs(x2 - x1), abs(y2 - y1)
            append_str_temp = (
                    str(round((x1 + x2) / 2 / scale_x, dot)) + ' ' +
                    str(round((y1 + y2) / 2 / scale_y, dot)) + ' ' +
                    str(round((w1 / scale_x), dot)) + ' ' +
                    str(round((h1 / scale_y), dot)) + ' ' +
                    str(current_class) + '\n'
            )
            cv2.putText(
                img,
                f'Selected Class: {current_class} (Press A/D to switch class)',
                (0, img.shape[0] // 2 + 30),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.7, (100, 255, 255), 2
            )

        elif key_insert is not None:
            position, Mouse_move, box_move = None, None, None
            cv2.putText(
                img,
                'Switching visibility: 0     1    2 (Press A/D to switch class)',
                (0, img.shape[0] // 2 - 30),
                cv2.FONT_HERSHEY_SIMPLEX,
                1, (100, 255, 255), 2,
                lineType=cv2.LINE_AA
            )
            # Handle class switching on key press
            if key_insert == 0:
                key_v[label_index_pos] = (key_v[label_index_pos] + 1) % 3
            elif key_insert == 1:
                key_v[label_index_pos] = (key_v[label_index_pos] + 1) % 3
            elif key_insert == 2:
                key_v[label_index_pos] = (key_v[label_index_pos] + 1) % 3
            
        else:
            img = img_s.copy()
            if Mouse_insert is not None:
                position, Mouse_move = None, None
                cv2.putText(
                    img,
                    'delete: W, exit: E, switch class: S (Press A/D to switch class)',
                    (0, img.shape[0] // 2 - 30),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    1, (100, 255, 255), 2,
                    lineType=cv2.LINE_AA
                )
            label_show(img, label_path, Mouse_insert)

        cv2.imshow('image', img)

        key = cv2.waitKey(1)

        if key == ord('Q'):
            break
        elif key == ord('E'):
            position, Mouse_move, box_move, end_draw_rectangle, append_str_temp = None, None, None, None, None
            if draw_rectangle is not None:
                draw_rectangle = None
            else:
                try:
                    label.pop()
                except IndexError:
                    pass
        elif key == ord('W'):
            position, Mouse_move, box_move, end_draw_rectangle, append_str_temp = None, None, None, None, None
            if draw_rectangle is not None:
                draw_rectangle = None
            else:
                try:
                    label.pop()
                except IndexError:
                    pass
#         elif key == ord('S') and end_draw_rectangle is None:
#             if key_insert is not None:
#                 if end_draw_rectangle is None:
#                     end_draw_rectangle = True
#                 else:
#                     end_draw_rectangle = None
#             else:
#                 switch_class()
        elif key == ord('S') and end_draw_rectangle is None:
            if key_insert is not None:
                key_v[label_index_pos] = (key_v[label_index_pos] + 1) % 3
            else:
                switch_class()
                #cv2.setTrackbarPos('Class', 'image', current_class)
                img = img_s.copy()
                cv2.imshow('image', img)
        
        elif key == ord('D') and end_draw_rectangle is None:
            current_class += 1
            if current_class >= len(my_cls):
                current_class = 0
            cv2.setTrackbarPos('Class', 'image', current_class)
        elif key == ord('A') and end_draw_rectangle is None:
            current_class -= 1
            if current_class < 0:
                current_class = len(my_cls) - 1
            cv2.setTrackbarPos('Class', 'image', current_class)

    cv2.destroyAllWindows()    

if __name__ == '__main__':
    global label_path_main, image_path_main
    image_files = os.listdir(image_path)
    label_path_main = label_path  # Save the main label path
    for im in image_files:
        flag_init()
        im_path = os.path.join(image_path, im)
        la_path = os.path.join(label_path, im.split('.')[0] + '.txt')
        if key_point_is:
            create_file_key(im_path, la_path)
        main(im_path, la_path)    
```
